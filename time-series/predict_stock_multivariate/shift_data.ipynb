```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('apple.csv', header=0)
```

```python
print('df.columns :\n', list(df.columns))
MAX_PRICE = df.High.max()
MAX_VOL = df.Volume.max()
print('MAX_PRICE : ', MAX_PRICE)
print('MAX_VOL : ', MAX_VOL)

df.Volume = df.Volume / MAX_VOL
df.Open = df.Open / MAX_PRICE
df.Close = df.Close / MAX_PRICE
df.High = df.High / MAX_PRICE
df.Low = df.Low / MAX_PRICE
```

```python
dff = df[:20]
dff
```

```python
shift_days = 5
y_label = 'High'
```

```python
# shift
dff[y_label].index = dff[y_label].index - shift_days
_dff = dff.drop(y_label, axis=1)
```

```python
# shift 된 데이터 셋 
a = pd.concat([_dff, dff[y_label]], axis=1)
a
```

```python
# 앞 뒤 shift 된 데이터를 제외한 학습 셋 
shift_df = a[shift_days:len(a)-shift_days]
shift_df
```

```python
# target 값이 없는 미래 예측에 활용될 데이터 
future_pred_data = a[-shift_days:]
future_pred_data
```
